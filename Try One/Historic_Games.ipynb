```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

We are starting with the 2021 Games since they are the most recent. We are only looking at medal wins so far. 

```python
#import data
data = pd.read_excel('2021_Games/Medals.xlsx', index_col=0)
print(data)
```

```python
barWidth = 0.25

teams = data.loc[: ,"Team/NOC"] #Get the teams
golds = data.loc[: ,"Gold"]
silvers = data.loc[: ,"Silver"]
bronzes = data.loc[: ,"Bronze"]

br1 = np.arange(len(golds)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, golds, color = 'gold', width = barWidth, label = 'Golds')
plt.bar(br2, silvers, color = 'silver', width = barWidth, label = 'Silvers')
plt.bar(br3, bronzes, color = 'brown', width = barWidth, label = 'Bronzes')

plt.xlabel("Teams")
plt.ylabel("Medal Count")
plt.title("Medals for each team")

plt.legend()
plt.show()
```

```python
barWidth = 0.25

x = 10 #Number of top teams we want to grab

teams = teams.head(x) #Get the top x teams
golds = golds.head(x)
silvers = silvers.head(x)
bronzes = bronzes.head(x)

br1 = np.arange(len(golds)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, golds, color = 'gold', width = barWidth, label = 'Golds')
plt.bar(br2, silvers, color = 'silver', width = barWidth, label = 'Silvers')
plt.bar(br3, bronzes, color = 'brown', width = barWidth, label = 'Bronzes')

plt.title("Medals for top 10 Countries")
plt.xticks([r + barWidth for r in range(len(golds))], teams, rotation=-45)


plt.legend()
plt.show()
```

```python
#Get the percent of medals won by team/country

teams = data.loc[: ,"Team/NOC"]
golds = data.loc[: ,"Gold"]
silvers = data.loc[: ,"Silver"]
bronzes = data.loc[: ,"Bronze"]
total = data.loc[: ,"Total"]

total_medals = np.sum(total)
total_golds = np.sum(golds)
total_silvers = np.sum(silvers)
total_bronzes = np.sum(bronzes)

total_percent = []
gold_percent = []
silver_percent = []
bronze_percent = []
for num in range(len(total)):
    total_percent.append(round(((total.iloc[num] / total_medals) * 100), 2))
    gold_percent.append(round(((golds.iloc[num] / total_golds) * 100), 2))
    silver_percent.append(round(((silvers.iloc[num] / total_silvers) * 100), 2))
    bronze_percent.append(round(((bronzes.iloc[num] / total_bronzes) * 100), 2))
percents = {'Team': teams, 'Golds': gold_percent, 'Silvers': silver_percent, 'Bronzes': bronze_percent, 'Total': total_percent}
percents_df = pd.DataFrame(data = percents)
print(percents_df)
```

This shows us the percent of medals each country won for total medals and for each diffrent type of medales.
